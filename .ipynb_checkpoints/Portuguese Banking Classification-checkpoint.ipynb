## Predicting Marketing Customers in Portuguese Bank

```python
## Import Packages 
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, brier_score_loss
from sklearn.metrics import matthews_corrcoef, classification_report, confusion_matrix,log_loss 
from sklearn.metrics import plot_confusion_matrix, roc_auc_score, roc_curve,precision_recall_curve, auc
from sklearn.metrics import ConfusionMatrixDisplay

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
```

```python
## We preview our banking dataset
df = pd.read_csv(r'C:\Users\benja\OneDrive\Desktop\Datasets\bank-additional-full.csv')
df.head()
```

```python
## We look at the columns 
df.columns
```

```python
## We get info
df.info()
```

```python
## We get quick summary stats
df.describe()
```

```python
## Check for null values
df.isnull().sum()
```

```python
## Change category columns to categorical
categorical_columns = ['job','education','marital']
for i in categorical_columns:
    df[i] = df[i].astype('category')
```

```python
## Plot Initial Heatmap for Quick visual EDA
plt.figure(figsize=(14,7))
sns.heatmap(df.corr(), annot = True, mask = np.triu(df.corr()))
plt.title('Correlation Heatmap', pad = 20, fontsize = 20)
plt.show()
```

# Preprocessing:

### 1) One-Hot Encoding - Getting Dummies

```python
## Get all categorical / object columns: 
categorical_cols = list(df.select_dtypes(['category','object']).columns)
```

```python
## Create dummies out of the categorical / object columns. 
dummies = pd.get_dummies(df[categorical_cols], drop_first = True)
dummies.head()
```

```python
## Remove the original non-dummied columns from df
df.drop(columns =categorical_cols, inplace = True)
```

```python
## Combine the new dummie columns with the df 
df = pd.concat([df,dummies], axis =1)
```

```python
## Print a new heatmap
plt.figure(figsize=(40,20))
sns.heatmap(df.corr(), annot = True, mask = np.triu(df.corr()))
plt.title('Correlation Heatmap', pad = 20, fontsize = 20)
plt.show()
```

### 2) Feature Scaling: 

```python
## Feature scaling with StandardScaler. Initialize the columns to scaler and define columns to scale.
scaler = StandardScaler()
cols_scale = ['age','duration','pdays','cons.price.idx','cons.conf.idx','euribor3m','nr.employed']
```

```python
## Fit and transform the new columns, fit them in a a new dataframe. 
X_scaled = scaler.fit_transform(df[cols_scale])
scaled_categories = pd.DataFrame(X_scaled, columns = cols_scale)
```

```python
## Remove original columns and add back the new scaled columns to the df. 
df.drop(columns=cols_scale, inplace = True)
df = pd.concat([scaled_categories,df], axis =1)
df.head()
```

### 3)  Variable Definition and Train_test_split

```python
## Variable definitions
X = df.drop(columns = ['y_yes'])
y = df.y_yes
print(X.shape, y.shape)
```

```python
## Split into training and test sets.
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25,random_state =42)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
## Check distributions of the class 
proportion = (sum(y) / len(y))
print(y.value_counts())
print("\n")
print('Positive Class Proportion: ' +str(round(proportion,3)))
```

# Modelling:

## Scenario 1: Testing of three models: 
- logreg
- rfc
- ada 

```python
## Instantiate, fit, predict Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_test)
y_pred_thresh_30 = (logreg.predict_proba(X_test)[:,1] >=0.37).astype(int)
y_pred_proba = logreg.predict_proba(X_test)[:,1]
```

```python
## Instantiate, fit, predict Random Forrest
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
y_pred_rfc = rfc.predict(X_test)
y_pred_thresh_30_rfc = (rfc.predict_proba(X_test)[:,1] >=0.3).astype(int)
y_pred_proba_rfc = rfc.predict_proba(X_test)[:,1]
```

```python
## Instantiate, fit and predict AdaBoost 
ada = AdaBoostClassifier()
ada.fit(X_train,y_train)
y_pred_ada = ada.predict(X_test)
y_pred_ada_threshold_30 = (ada.predict_proba(X_test)[:,1] >=0.3).astype(int)
y_pred_proba_ada = ada.predict_proba(X_test)[:,1]
```

```python
## Print out metrics for Random Forest
print('Metrics:')
print('Accuracy Score: ' + str(round(accuracy_score(y_test,y_pred),4)))
print('Accuracy Score at Threshold 37%: '+ str(round(accuracy_score(y_test,y_pred_thresh_30),4)))
print('Precision Score: ' + str(round(precision_score(y_test,y_pred),3)))
print('Recall Score: ' + str(round(recall_score(y_test,y_pred),3)))
print('F1_Score Score: ' + str(round(f1_score(y_test,y_pred),3)))
print('ROC_AUC Score: ' + str(round(roc_auc_score(y_test,y_pred_proba),3)))
print('Log Loss Score: ' + str(round(log_loss(y_test,y_pred),3)))
print('MCC: ' + str(round(matthews_corrcoef(y_test,y_pred),3)))
print("\n\n Classification Report")
print(classification_report(y_test,y_pred))
print("\n\n Confusion Matrix")
print(confusion_matrix(y_test, y_pred))
print("\n\n")
print("First 50 actuals: ")
print(np.array(y_test[0:50]))
print("\n First 50 preds: ")
print(y_pred[0:50])
print("\n First 50 preds at threshold 0.37: ")
print(y_pred_thresh_30[0:50])
print("\n First 50 pred probabilities : ")
print(y_pred_proba[0:50])
print("\n")
plot_confusion_matrix(logreg,X_test,y_test)
plt.show()
```

## ROC Curves
A ROC curve is a diagnostic plot that evaluates a set of probability predictions made by a model on a test dataset.

```python
## Plotting the ROC curve and finding the best Threshold. 
fpr,tpr, threshold = roc_curve(y_test,y_pred_proba)
fpr_rfc, tpr_rfc , threshold_rfc = roc_curve(y_test,y_pred_proba_rfc)
fpr_ada, tpr_ada, threshold_ada  = roc_curve(y_test, y_pred_proba_ada)

plt.figure(figsize=(20,10))

## Plotting No Skill 
plt.plot([0,1], [0,1], linestyle='--', label='No Skill', color = 'black')

## Plotting Logistic Regression: 
plt.plot(fpr, tpr, c='lawngreen',
         label='Logistic AUC: ' + str(round(roc_auc_score(y_test,y_pred_proba),3)))

## Plotting RandomForest: 
plt.plot(fpr_rfc, tpr_rfc, marker='.', c='red', 
         label='Random_Forest AUC: ' +str(round(roc_auc_score(y_test,y_pred_proba_rfc),3)))

## Plotting AdaBoost:
plt.plot(fpr_ada, tpr_ada, c='dodgerblue', linestyle ='--',
         label='AdaBoost AUC: ' +str(round(roc_auc_score(y_test,y_pred_proba_ada),3)))

## Optional but this is the best threshold for logistic regression
gmeans = np.sqrt(tpr * (1-fpr))
ix = np.argmax(gmeans)
plt.scatter(fpr[ix], tpr[ix], marker='o', color='black', label='Best LogReg')
print('Best Threshold=%f, G-Mean=%.3f' % (threshold[ix], gmeans[ix]))


## Labelling 
plt.xlabel('False Positive Rate', fontsize =15)
plt.ylabel('True Positive Rate', fontsize =15)
plt.title('ROC Curves for each Basic Model ', pad =25, fontsize = 20)
plt.legend()
plt.show()
```

## Precision-Recall Curves 
Unlike the ROC Curve, a precision-recall curve focuses on the performance of a classifier on the positive (minority class) only.

```python
## Plotting Precision Recall Curves for our three basic models. 

# Start by defining the precision and recalls for each at each threshold and defining no skill. 
precision, recall, thresholds = precision_recall_curve(y_test,y_pred_proba)
precision_rfc , recall_rfc, thresholds_rfc = precision_recall_curve(y_test, y_pred_proba_rfc)
precision_ada , recall_ada, thresholds_ada = precision_recall_curve(y_test, y_pred_proba_ada)
no_skill = len(y_test[y_test==1]) / len(y_test)

## Plot No Skill
plt.figure(figsize=(20,10))
plt.plot([0,1], [no_skill,no_skill], linestyle='--', label='No Skill')



## Precision - Recall Curve for Logistic Regression 
plt.plot(recall, precision, marker='.', c = 'lawngreen',
         label='Logistic' + str(round(auc(recall,precision),3)))

## Precision - Recall Curve for Random Forest 
plt.plot(recall_rfc, precision_rfc, marker='.', c= 'red',
         label='Random Forest: ' + str(round(auc(recall_rfc,precision_rfc),3)))

## Precision - Recall Curve for Adaboost 
plt.plot(recall_ada, precision_ada, marker='.', c = 'dodgerblue',
         label='AdaBoost: ' + str(round(auc(recall_ada,precision_ada),3)))


## Labelling of axes. 
plt.xlabel('Recall', fontsize = 15)
plt.ylabel('Precision', fontsize = 15)
plt.title('Precision / Recall Curve for three models', fontsize = 20, pad = 20)
plt.annotate('No Skill Line: '+str(round(no_skill,3)),(0.2,0.12), fontsize =15)
plt.legend()
plt.show()
```

## Evaluation

```python
## Getting the optimal thresholds for each model and printing out the classification report for the best score. 
def to_labels(pos_probs, threshold):
    return (pos_probs >= threshold).astype('int')

thresholds = np.arange(0, 1, 0.001)

probs_log = logreg.predict_proba(X_test)[:,1]
rfc_log = rfc.predict_proba(X_test)[:,1]
ada_log = ada.predict_proba(X_test)[:,1]

log_scores = [f1_score(y_test,to_labels(probs_log, t)) for t in thresholds]
rfc_scores = [f1_score(y_test,to_labels(rfc_log, t)) for t in thresholds]
ada_scores = [f1_score(y_test,to_labels(ada_log,t)) for t in thresholds]

ix_log = np.argmax(log_scores)
ix_rfc = np.argmax(rfc_scores)
ix_ada = np.argmax(ada_scores)

print('Log Best Threshold=%.3f, Best F-Score=%.5f' % (thresholds[ix_log], log_scores[ix_log]))
print('RFC Best Threshold=%.3f, Best F-Score=%.5f' % (thresholds[ix_rfc], rfc_scores[ix_rfc]))
print('ADA Best Threshold=%.3f, Best F-Score=%.5f' % (thresholds[ix_ada], ada_scores[ix_ada]))

## We get the classification report
print("\n ADABoost Classification Report: \n")
print(classification_report(y_test,(ada.predict_proba(X_test)[:,1] >=0.493).astype(int)))

## We get the confusion matrix for the best threshold of the best model. 
print("\n ADABoost Confusion Matrix: \n")
matrix = confusion_matrix(y_test,(ada.predict_proba(X_test)[:,1] >= 0.493).astype(int), labels = ada.classes_)
disp = ConfusionMatrixDisplay(matrix, display_labels=ada.classes_)
disp.plot()
plt.show()
```

```python
## RFC 
rfc_grid={'n_estimators': [20,50,100,200,500],'max_features': ['auto', 'sqrt', 'log2'],'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']}

rfc = RandomForestClassifier()
rfc_cv = RandomizedSearchCV(rfc, rfc_grid , cv=10, scoring='precision', n_jobs=1)
rfc_cv.fit(X_train,y_train)

print("RFC Tuned hyperparameters :(best parameters) ", rfc_cv.best_params_)
print("RFC Best score :", rfc_cv.best_score_)
```

```python
## AdaBoost 
ada_grid={'n_estimators': [10,50,100,500], 'learning_rate': [0.0001,0.001,0.01,0.1,1], 'algorithm':['SAMME','SAMME.R']}
ada = AdaBoostClassifier()
ada_cv = RandomizedSearchCV(ada, ada_grid , cv=10, scoring='precision', n_jobs =-1)
ada_cv.fit(X_train,y_train)

print("Ada Tuned hyperparameters :(best parameters) ", ada_cv.best_params_)
print("Ada Best score :", ada_cv.best_score_)
```
