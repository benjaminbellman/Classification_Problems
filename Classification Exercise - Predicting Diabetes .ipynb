## Classification: Predicting Diabetes

```python
## We import packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, r2_score, plot_confusion_matrix, precision_score, recall_score, roc_curve,auc
from sklearn.metrics import roc_auc_score, RocCurveDisplay, precision_recall_curve,f1_score


import warnings 
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
## We load the dataset and preview 
diabetes = pd.read_csv(r'C:\Users\benja\OneDrive\Desktop\Datasets\Diabetes.csv')
diabetes.head()
```

```python
## We look at the columns:
diabetes.info()
```

```python
## Check for null values: 
diabetes.isna().sum()
```

```python
## Let's look at the distribution of our target variable: 
print(diabetes.Outcome.value_counts())
print("\n")
mean_diabetes = round((diabetes[diabetes.Outcome == 1]['Outcome']).count() / len(diabetes.Outcome),2)
print('Proportion wbmean_diabetes: ' + str(mean_diabetes))
```

```python
## Quick heatmap to visualize correlations: 
plt.figure(figsize=(15,8))
sns.heatmap(diabetes.corr(), annot=True, annot_kws={"size": 16}, mask = np.triu(diabetes.corr()) )
plt.title('Feature Variable Correlation with Outcome', pad =25, fontsize = 25)
plt.show()
```

```python
## We seperate our data in feature and target variables and split them into training and test groups 
X = diabetes.drop(columns='Outcome')
y = diabetes.Outcome
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled.shape,y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size = 0.25, random_state = 4)
```

## We will use 7 different Classification Algorithms: 

- 1) Logistic Regression
- 2) RandomForrest
- 3) Support Vector Machines
- 4) Gradient Boosting 
- 5) Naive Bayes
- 6) K-NN
- 7) Stochastic Gradient Descent 

## 1) Logistic Regression

```python
## We instantiate Logistic Regression Model and fit it to training data, then predict on the test data. 
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
y_pred_log = log_reg.predict(X_test)
accuracy_log  = accuracy_score(y_test,y_pred_log)
precision_log = precision_score(y_test,y_pred_log)
recall_log = recall_score(y_test,y_pred_log)
print("Actual Data: " +str(np.array(y_test)))
print("\n")
print("Model Predictions: " +str(y_pred_log))
print("\n")
print("Accuracy Score: " + str(accuracy_log))
print("Precision Score: " +str(precision_log) )
print("Recall Score: " +str(recall_log) )
```

```python
plot_confusion_matrix(log_reg, X_test,y_test)
plt.title('Confusion Matrix of Test Data', pad =25)
plt.show()
```

```python
## We instantiate a logistic regression model and use GridSearchCV to find the best parameters. 
log_reg = LogisticRegression()
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
             'l1_ratio': [0,.5,1],
             'penalty':['none','l1','l2','elasticnet']
             }
lr_cv = GridSearchCV(log_reg, param_grid, cv=10, scoring='roc_auc')
lr_cv.fit(X_train,y_train)

## We print the metrics: 
print('Best Estimator:' +str(lr_cv.best_estimator_))
print('Best Params:' +str(lr_cv.best_params_))
print('Best Score:' +str(lr_cv.best_score_))
```

```python
## We refit Logistic Regression with the better parameters found from GridSearchCV
log_reg = LogisticRegression(C=0.1, l1_ratio=0, penalty='l2')
log_reg.fit(X_train,y_train)
y_pred_log = log_reg.predict(X_test)
accuracy_log  = accuracy_score(y_test,y_pred_log)
precision_log = precision_score(y_test,y_pred_log)
recall_log = recall_score(y_test,y_pred_log)
print("Actual Data: " +str(np.array(y_test)))
print("\n")
print("Model Predictions: " +str(y_pred_log))
print("\n")
print("Accuracy Score: " + str(accuracy_log))
print("Precision Score: " +str(precision_log) )
print("Recall Score: " +str(recall_log) )
```

```python
plot_confusion_matrix(log_reg, X_test,y_test)
plt.title('Confusion Matrix of Test Data', pad =25)
plt.show()
```

## 2) RandomForrestClassifier 

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred_rf = rf.predict(X_test)
accuracy_rf = accuracy_score(y_test,y_pred_rf)
precision_rf = precision_score(y_test,y_pred_rf)
recall_rf = recall_score(y_test,y_pred_rf)
print("Actual Data: " + str(np.array(y_test)))
print("\n")
print("Model Predictions: " +str(y_pred_rf))
print("\n")
print("Accuracy Score: " +str(accuracy_rf) )
print("Precision Score: " +str(precision_rf) )
print("Recall Score: " +str(recall_rf) )
```

```python
plot_confusion_matrix(rf, X_test,y_test)
plt.title('Confusion Matrix of Test Data', pad =25)
plt.show()
```

```python
## We use GridSearchCV to r
rfc = RandomForestClassifier(random_state = 42)
param_grid = { 
    'n_estimators': [10, 25, 50, 100, 200],
    'max_features': ['sqrt', 'auto', 'log2'],
    'criterion': ['gini','entropy'],
    'max_depth' : [2,4,6,8],
    'bootstrap':['False', 'True']
}

rfc_GS = GridSearchCV(rfc, param_grid, cv=5, scoring='roc_auc')
rfc_GS.fit(X_train,y_train)

##Let's look at the results  for Grid Search CV 
print('Best Estimator:' +str(rfc_GS.best_estimator_))
print('Best Params:' +str(rfc_GS.best_params_))
print('Best Score:' +str(rfc_GS.best_score_))
```

```python
## With the best parameters, we re-instantiate the model, fit and predict the probabilities. 
rfc_best = RandomForestClassifier(bootstrap= False, criterion= 'gini', max_depth= 4, 
                                  max_features= 'sqrt', n_estimators= 50, random_state = 42)
rfc_best.fit(X_train, y_train)
y_pred_best_rfc = rfc_best.predict(X_test)
```

```python
## We define our metrics and print them out.
accuracy_best_rfc  = accuracy_score(y_test,y_pred_best_rfc)
precision_best_rfc = precision_score(y_test,y_pred_best_rfc)
recall_best_rfc = recall_score(y_test,y_pred_best_rfc)
f1_best_rfc = f1_score(y_test,y_pred_best_rfc)
print("Accuracy Score: " + str(accuracy_best_rfc ))
print("Precision Score: " +str(precision_best_rfc))
print("Recall Score: " +str(recall_best_rfc))
print("F1 Score: " +str(f1_best_rfc))

## We plot the confusion Matrix to visualize the predictions
plot_confusion_matrix(rfc_best, X_test,y_test)
plt.title('Confusion Matrix of Test Data', pad =25)
plt.show()
```

```python
## Plotting the Precision / Recall Tradeoff 

names = ["Logistic Regression","SVC"]
models = [log_reg,svc]
colors = ['r','m']
linestyles =['-',':']

plt.figure(figsize=(15,10))
counter = 0
for name, model in zip(names,models):
    y_decision_function = model.decision_function(X_test)
    precision,recall,threshold = precision_recall_curve(y_test,y_decision_function)
    plt.plot(recall,precision,color=colors[counter], linestyle=linestyles[counter], label = name)
    counter +=1
    
plt.title('Precision Recall Tradeoff', pad=20, fontsize = 20)
plt.xlabel('Recall', fontsize = 15)
plt.ylabel('Precision', fontsize = 15)
plt.legend()
plt.show()
```
