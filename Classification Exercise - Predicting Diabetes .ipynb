## Classification: Predicting Diabetes

```python
## We import packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, r2_score, plot_confusion_matrix, precision_score, recall_score, roc_curve,auc
from sklearn.metrics import roc_auc_score, RocCurveDisplay
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
import warnings 
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
## We load the dataset and preview 
diabetes = pd.read_csv(r'C:\Users\benja\OneDrive\Desktop\Datasets\Diabetes.csv')
diabetes.head()
```

```python
## We look at the columns:
diabetes.info()
```

```python
## Check for null values: 
diabetes.isna().sum()
```

```python
## Let's look at the distribution of our target variable: 
print(diabetes.Outcome.value_counts())
print("\n")
mean_diabetes = round((diabetes[diabetes.Outcome == 1]['Outcome']).count() / len(diabetes.Outcome),2)
print('Proportion wbmean_diabetes: ' + str(mean_diabetes))
```

```python
## Quick heatmap to visualize correlations: 
plt.figure(figsize=(15,8))
sns.heatmap(diabetes.corr(), annot=True, annot_kws={"size": 16}, mask = np.triu(diabetes.corr()) )
plt.title('Feature Variable Correlation with Outcome', pad =25, fontsize = 25)
plt.show()
```

```python
## We seperate our data in feature and target variables and split them into training and test groups 
X = diabetes.drop(columns='Outcome')
y = diabetes.Outcome
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled.shape,y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size = 0.25, random_state = 4)
```

## 1) Logistic Regression

```python
## We instantiate Logistic Regression Model and fit it to training data, then predict on the test data. 
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
y_pred_log = log_reg.predict(X_test)
accuracy_log  = accuracy_score(y_test,y_pred_log)
precision_log = precision_score(y_test,y_pred_log)
recall_log = recall_score(y_test,y_pred_log)
print("Actual Data: " +str(np.array(y_test)))
print("\n")
print("Model Predictions: " +str(y_pred_log))
print("\n")
print("Accuracy Score: " + str(accuracy_log))
print("Precision Score: " +str(precision_log) )
print("Recall Score: " +str(recall_log) )
```

```python
plot_confusion_matrix(log_reg, X_test,y_test)
plt.title('Confusion Matrix of Test Data', pad =25)
plt.show()
```

## 2) RandomForrestClassifier 

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred_rf = rf.predict(X_test)
accuracy_rf = accuracy_score(y_test,y_pred_rf)
precision_rf = precision_score(y_test,y_pred_rf)
recall_rf = recall_score(y_test,y_pred_rf)
print("Actual Data: " + str(np.array(y_test)))
print("\n")
print("Model Predictions: " +str(y_pred_rf))
print("\n")
print("Accuracy Score: " +str(accuracy_rf) )
print("Precision Score: " +str(precision_rf) )
print("Recall Score: " +str(recall_rf) )
```

```python
plot_confusion_matrix(rf, X_test,y_test)
plt.title('Confusion Matrix of Test Data', pad =25)
plt.show()
```

```python
## We can get the importance of the feature variables in the Random Forest Model. 
features = pd.DataFrame(rf.feature_importances_, index = X.columns, columns=['Feature Importance']).sort_values('Feature Importance',ascending=False)
features
```

```python
## We can even visualize and plot the feature importance: 
sns.color_palette("magma", as_cmap=True)
sns.barplot(y = features.index, x = features['Feature Importance'],palette='magma')
plt.title('Feature importance in the Random Forest Model', pad = 20, fontsize = 15)
plt.show()
```

## 3) Support Vector Machines

```python
svc = SVC(probability = True)
svc.fit(X_train,y_train)
y_pred_svc = svc.predict(X_test)
accuracy_svc = accuracy_score(y_test,y_pred_svc )
precision_svc = precision_score(y_test,y_pred_svc )
recall_svc = recall_score(y_test,y_pred_svc )
print("Actual Data: " + str(np.array(y_test)))
print("\n")
print("Model Predictions: " +str(y_pred_svc ))
print("\n")
print("Accuracy Score: " +str(accuracy_svc) )
print("Precision Score: " +str(precision_svc) )
print("Recall Score: " +str(recall_svc) )
```

```python
plot_confusion_matrix(svc, X_test,y_test)
plt.title('Confusion Matrix of Test Data', pad =25)
plt.show()
```

## Plot ROC and AUC

#### Important Note: 
- Above in the y_pred, we are making a prediction on whether the person has diabetes or not (1 or 0)
- To plot the ROC and AUC, we need to use **'decision_function' method intead of 'predict'**!
- What it outputs is a number that tells how far the point is from the classification plane. It returns the distance from that plane. 

```python
## We define them here: 
y_predict_log = log_reg.decision_function(X_test)
y_predict_svc = svc.decision_function(X_test)

## Random Forest does not have a decision function. We do this : 
rf_disp = RocCurveDisplay.from_estimator(rf, X_test, y_test, alpha=0.8)
```
